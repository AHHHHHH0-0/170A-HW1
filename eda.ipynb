```python
from pathlib import Path
import pandas as pd 

names = [
    'acm_fellows', 
    'conference_ranking', 
    'country_info', 
    'csrankings', 
    'data', 
    'dblp_alias', 
    'field_conference', 
    'generated_author_info', 
    'geo_location', 
    'turing'
]

# Load files
dir = Path('data/')
files = {name:pd.read_csv(file) for name, file in zip(names, dir.iterdir())}
```

```python
# General information
for f in names: 
    file = files[f]

    print('='*100)
    print(f, file.shape)
    print('='*100)
    print(file.head(3))
    print('-'*100)
    print(file.describe())
    print('-'*100)
    print(file.isnull().sum())
    print()

```
