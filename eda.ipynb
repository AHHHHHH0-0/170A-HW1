```python
from pathlib import Path
import pandas as pd 

# Load files
dir = Path('data/')
files = {file.name:pd.read_csv(file) for file in dir.iterdir()}
```

```python
# General information
for f in files: 
    file = files[f]

    print('='*100)
    print(f, file.shape)
    print('='*100)
    print(file.dtypes)
    print('-'*100)
    print(file.head(10))
    print(file.tail(10))
    print('-'*100)
    print(file.describe())
    print('-'*100)
    print(file.isnull().sum())
    print()

```
